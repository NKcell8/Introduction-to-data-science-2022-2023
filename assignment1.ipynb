## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:

new_df = df.loc[(df['bedrooms']==3) & (df['bathrooms']==2) & (df['floors']==2)]
new_df.describe()
#Hint: pandas.DataFrame.loc
# I got basic statistics of above constraints.
```

```python
#b)
#Your code:
figure(figsize=(10, 8.5), dpi=80)
plt.scatter(df['grade'],df['price'],c='#000000',s=15)
plt.scatter(new_df['grade'],new_df['price'],c='#ffef00',s=40,edgecolors='#000000')
plt.xlabel('grade')
plt.ylabel('price')
plt.show()

# I can see on the plot that chosen houses have grade between 6 and 10.
```

```python
#c)
#Your code:

priceToFootageRatio = []

for i in range(0,len(new_df.index)):
  result = new_df['price'].values[i]/new_df["sqft_living"].values[i]
  priceToFootageRatio.append(result)

new_df["ratio_price_sqft_living"] = priceToFootageRatio
new_df.sort_values(by="ratio_price_sqft_living").head()



#Hint: .loc, .sort_values(), .index

# House with the best price to footage ratio costs 260 000 and space for living is 2710
```
